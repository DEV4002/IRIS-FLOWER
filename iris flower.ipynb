```python
# convert the last column to  numeric
import pandas as pd
df= pd.read_csv("C:\\Users\\asus\\OneDrive\\Desktop\\PYTHONSOFT\\IRIS FLOWER PROJECT\\iris\\iris.csv",names=['sl','sw','pl','pw','class'])

df['class']=df['class'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df.sample(5)
# separate input and output
X=df.drop(columns=['class'])
Y=df['class']
# split the data set for training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=5)
X_train
#######################################################################################
import warnings
warnings.filterwarnings('ignore')
import tkinter.messagebox as m
def knn():
    global acc_knn
    from sklearn.neighbors import KNeighborsClassifier
    K=KNeighborsClassifier(n_neighbors=5)
    # train the modell by using 120 flower data
    K.fit(X_train,Y_train)
    Y_pred_knn=K.predict(X_test)
    Y_pred_knn
    #find accurecy of knn model

    from sklearn.metrics import accuracy_score
    acc_knn=accuracy_score(Y_test,Y_pred_knn)
    acc_knn=round(acc_knn*100,2)
    print("accuracy in knn model is",acc_knn,"%")
    m.showinfo(title="KNN",message="accuracy is"+str(acc_knn)+"%")
    
def lg():
    global acc_lg
    global L
    # implement logistic regressionf
    from sklearn.linear_model import LogisticRegression 
    L=LogisticRegression()
    L.fit(X_train,Y_train)
    Y_pred_lg=L.predict(X_test)
    from sklearn.metrics import accuracy_score
    acc_lg=accuracy_score(Y_test,Y_pred_lg)
    acc_lg=round(acc_lg*100,2)
    print("accuracy in logistic reg model is",acc_lg,"%")
    m.showinfo(title="LG",message="accuracy is"+str(acc_lg)+"%")
        
def nb():
    global acc_nb
    from sklearn.naive_bayes import GaussianNB
    N=GaussianNB()
    
    N.fit(X_train,Y_train)
    Y_pred_nb=N.predict(X_test)
    from sklearn.metrics import accuracy_score
    acc_nb=accuracy_score(Y_test,Y_pred_nb)
    acc_nb=round(acc_nb*100,2)
    print("accuracy in decision treee model is ",acc_nb,"%")
    m.showinfo(title="NB",message="accuracy is"+str(acc_nb)+"%")
def dt():
    global acc_dt
    # implement Decision tree model
    from sklearn.tree import DecisionTreeClassifier
    D=DecisionTreeClassifier()
    D.fit(X_train,Y_train)
    Y_pred_dt=D.predict(X_test)
    from sklearn.metrics import accuracy_score
    acc_dt=accuracy_score(Y_test,Y_pred_dt)
    acc_dt=round(acc_dt*100,2)
    print("accuracy in decision treee model is ",acc_dt,"%")
    m.showinfo(title="DT",message="accuracy is"+str(acc_dt)+"%")
def compare():
    import matplotlib.pyplot as plt
    models=['KNN','LG','DT','NB']
    accuracy=[acc_knn,acc_lg,acc_dt,acc_nb]
    plt.bar(models,accuracy ,color=['yellow','green','red','blue'])
    plt.xlabel("MODELS")
    plt.ylabel("ACCURAcy")
    plt.show()
    m.showinfo(title="BEST MODEL",message="LOGISTIC REGRESSION")
def SUBMIT():
    lg()
    sl=float(s1.get())
    pl=float(s3.get())
    sw=float(s2.get())
    pw=float(s4.get())
    result=L.predict([[sl,sw,pl,pw]])
    if result[0]==0:
        flower='SETOSA'
    elif result[0]==1:
        flower="VERSICOLOR"
    else :
        flower ='VERGINICIA'
    m.showinfo(title="IRIS PROJECT",message=flower)
    
    
def RESET():
    s1.delete(0,END)
    s2.delete(0,END)
    s3.delete(0,END)
    s4.delete(0,END)
    
########################################################################################################################################################
from tkinter import *

w=Tk()
w.title("IRIS PROJECT")
L1=Label(w,text="IRIS FLOWER PROJECT ",font=('arial',20,'bold'),fg="blue",borderwidth=5,relief="solid")
L1.grid(row=1,column=1,columnspan=4)

Bknn=Button(w,text="KNN",font=('arial',20,'bold'),fg="red",borderwidth=3,relief="raise",command=knn)
Blg=Button(w,text="LG",font=('arial',20,'bold'),fg="red",borderwidth=3,relief="raise",command=lg)
Bnb=Button(w,text="NB",font=('arial',20,'bold'),fg="red",borderwidth=3,relief="raise",command=nb)
Bdt=Button(w,text="DT",font=('arial',20,'bold'),fg="red",borderwidth=3,relief="raise",command=dt)
Bknn.grid(row=2,column=1)
Blg.grid(row=2,column=2)
Bnb.grid(row=2,column=3)
Bdt.grid(row=2,column=4)
##################
Bcon=Button(w,text="COMPARE",font=('arial',20,'bold'),fg="red",borderwidth=3,relief="raise",command=compare)
Bcon.grid(row=3,column=2,columnspan=2)

L2=Label(w,text="PREDICT FOR A NEW FLOWER ",font=('arial',20,'bold'),fg="green",borderwidth=3,relief="solid")
L2.grid(row=4,column=1,columnspan=4)

sl=Label(w,text="SL",font=('arial',20,'bold'),fg="green",borderwidth=3,relief="sunken")
sl.grid(row=5,column=1)
s1=Entry(w,font=('arial',15,'bold'),bg="pink",width=7,borderwidth=5)
s1.grid(row=5,column=2)

sw=Label(w,text="SW",font=('arial',20,'bold'),fg="green",borderwidth=3,relief="sunken")
sw.grid(row=5,column=3)
s2=Entry(w,font=('arial',15,'bold'),bg="pink",width=7,borderwidth=5)
s2.grid(row=5,column=4)

pl=Label(w,text="PL",font=('arial',20,'bold'),fg="green",borderwidth=3,relief="sunken")
pl.grid(row=6,column=1)

s3=Entry(w,font=('arial',15,'bold'),bg="pink",width=7,borderwidth=5)
s3.grid(row=6,column=2)


pw=Label(w,text="PW",font=('arial',20,'bold'),fg="green",borderwidth=3,relief="sunken")
pw.grid(row=6,column=3)
s4=Entry(w,font=('arial',15,'bold'),bg="pink",width=7,borderwidth=5)
s4.grid(row=6,column=4)

Bsub=Button(w,text="SUBMIT",font=('arial',20,'bold'),fg="dark orange",borderwidth=5,relief="raise",command =SUBMIT)
Bsub.grid(row=7,column=1,columnspan=2)
Bres=Button(w,text= "RESET",font=('arial',20,'bold'),fg="dark orange",borderwidth=5,relief="raise",command=RESET)
Bres.grid(row=7,column=3,columnspan=2)

w.mainloop()
```
